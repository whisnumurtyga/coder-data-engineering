## 2. Buat perkalian matrik tanpa library numpy (manual)

1. Konsep dari operasi hitung untuk perkalian matriks adalah mengalikan elemen-elemen baris pada matriks pertama dengan elemen-elemen kolom pada matriks ke dua. <br>

2. Setiap anggota elemen matriks dikalikan dengan anggota elemen matriks lainnya sesuai urutan dan aturan yang berlaku pada perkalian matriks. <br>

3. Sehingga, perkalian matriks hanya bisa dilakukan untuk banyaknya kolom matriks pertama sama dengan banyaknya baris pada matriks kedua.


````python
A = [[2, 1, 2],
     [1, 2, 1]]

B = [[2, 1],
     [5, 5],
     [3, 2]]

# Apakah bisa dikalikan ?
# A = 2 x 3 dan B = 3 x 2

# hasil perkalian matriks = baris matriks A x kolom matriks B
result = [[0, 0],
          [0, 0]]

print("len A : " , len(A))
print("len B : ", len(B))
print("len B[0] : ", len(B[0]))



for i in range(len(A)):  #2
    for j in range(len(B[0])):  #2
        for k in range(len(B)):  #3
            result[i][j] += A[i][k] * B[k][j]
            
print("\nResult : ")            
for d in result:
    print(d)
````
