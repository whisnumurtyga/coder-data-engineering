### IMPORT LIB

```python
# Import Lib
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

### Mengambil Konten

```python
# Mengambil Konten Halaman Web
url = "https://www.espn.com/soccer/standings/_/league/uefa.champions"
response = requests.get(url)
content = response.content
```

### Baca HTML 

```python
# Membuat Objek BeautifulSoup
soup = BeautifulSoup(content, 'html.parser')
```

### Cari Table

```python
tables = soup.find_all('table')
tables
```

### Mendapatkan Nama Grup

```python
group_names = tables[0].select('td > span')
group_names
```

### Mendapatkan data team

```python
team_names = soup.select('td > div')
team_names

group_index = 0
count = 0
datas_team = []
data_team = 4
for tn in team_names :
    if count % 4 == 0 and count > 1:
        count = 0
        group_index += 1
    group = group_names[group_index].text
    position = tn.find('span', class_='team-position').text
    code = tn.find('img')['alt']
    name = tn.find('span', class_="hide-mobile").text
    count += 1
    # print(f"group: {group}\tposition: {position}\tcode: {code}\tName: {name}")
    datas_team.append([group,position, code, name])
    # print(tn)
```

### Mendapatkan stats poin

```python
stats_table = tables[1]
row = stats_table.select('tr.Table__TR td.Table__TD:not(.tar)')

lens = 0
count = 0
stats = []
temp_point = []
for r in row:
    temp_point.append(r.text)
    count += 1
    if count % 8 == 0 and count > 1:
        stats.append(temp_point)
        temp_point = []
        lens += 1

# Tambahan kondisi untuk menambahkan temp_point terakhir
if temp_point:
    stats.append(temp_point)
    lens += 1

stats
```

```python
datas_team
```

### Merger List data tim dan data statsnya

```python
result = []

for row1, row2 in zip(datas_team, stats):
    result.append(row1 + row2)

len(result)
result
```

### Convert ke dataframe

```python
df = pd.DataFrame(result)
df.head()
```

### Rename Columns

```python
df = df.rename(columns={0: 'Group', 1: 'Position', 2: 'Code', 3: 'Name', 4: 'Games Played', 5: 'Wins', 6: 'Draws', 7: 'Losses', 8: 'Goals_For', 9: 'Goals_Against', 10: 'Goal_Difference', 11: 'Points'})
df
```

### Save to CSV


```python
df.to_csv('uefa.csv', index=False)
```
