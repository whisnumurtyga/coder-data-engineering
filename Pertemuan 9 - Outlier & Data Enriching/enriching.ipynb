## **Apa itu data enriching**?
- **Data enriching**, juga dikenal sebagai **data enhancement**.
- **Adalah** Proses meningkatkan kualitas dan nilai data dengan menambahkan informasi tambahan atau memperluas detail yang ada. 
- **Tujuan** dari data enriching adalah untuk membuat data lebih berguna, lengkap, dan lebih mudah dimanfaatkan dalam analisis dan pengambilan keputusan.

## **Teknik Data Enriching**
1. **Penggabungan Data**: <br>
>ex: menggabungkan data pelanggan dengan data demografis untuk mendapatkan pemahaman yang lebih baik tentang profil pelanggan.
2. **Pemperkayaan Data**: <br>
>ex: menambahkan informasi geografis seperti koordinat lintang dan bujur ke data alamat pelanggan.
3. **Normalisasi Data**: <br>
>ex: Mengubah format tanggal menjadi format yang konsisten di seluruh dataset.
4. **Pemrosesan Bahasa Alami**: <br>
>ex: mengidentifikasi kata kunci atau topik utama dari teks ulasan pelanggan.
5. **Validasi Data**: <br>
>ex: memverifikasi nomor telepon atau alamat email untuk memastikan kebenarannya.

## **Manfaat**
- Data enriching dapat memberikan manfaat dalam berbagai konteks bisnis, seperti pemahaman pelanggan yang lebih baik, segmentasi pasar yang lebih akurat, personalisasi pengalaman pengguna, atau pengambilan keputusan yang lebih cerdas.
- Dengan data yang lebih kaya dan terperinci, organisasi dapat meningkatkan pemahaman mereka tentang pelanggan, tren pasar, atau situasi bisnis secara keseluruhan.
