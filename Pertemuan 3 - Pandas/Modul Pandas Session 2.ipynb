# Grouping and Sorting

What is the GroupBy function? Pandas' GroupBy is a powerful and versatile function in Python. It allows you to split your data into separate groups to perform computations for better analysis.

### DataFrame.groupby (by=None, axis=0, level=None, as_index=True, sort=True, group_keys=_NoDefault.no_default, squeeze=_NoDefault.no_default, observed=False, dropna=True)

```python
import pandas as pd
df = pd.read_excel("DataSet/Movies.xlsx")
# pd.set_option("display.max_rows", 10)
# pd.set_option("display.max_columns",15)
pd.set_option('display.max_columns', None)
# pd.set_option('display.max_rows', None)
```

```python
df.head()
```

## Single Indexing Groupby()

```python
df.rating.unique()
```

```python
df.groupby('rating').rating.count()
```

```python
df['rating'].value_counts()
```

```python
df.head()
```

```python
df.groupby('budget').budget.min()
```

```python
df.head()
```

```python
df.groupby(['genre']).runtime.agg([min, max])
```

```python
df.head(2)
```

```python
genre = df.groupby(['genre']).size().sort_values(ascending=False).reset_index()
genre_yang_ada = genre['genre'].tolist()
```

```python
df.head(3)
```

```python
df['revenue'] = df.apply(lambda revenue : revenue['gross'] - revenue['budget'], axis=1)
df
```

```python
negara = df['country'].value_counts().head(10).reset_index().rename(columns={'index': 'Negara','country':'jumlah'})
negara
```

```python
df.head()
```

# Sorting

```python
negara.sort_values(by='jumlah',ascending = True)
```

```python
genre = df.groupby(['country','genre'])['genre'].count().sort_values().to_frame().rename(columns={'genre':'jumlah'})
```

```python
genre = genre.sort_values(by=['jumlah'],ascending = False)
genre
```

```python
genre = genre.reset_index()
```

```python
genre
```

```python
genre.sort_index(ascending = False)
```
